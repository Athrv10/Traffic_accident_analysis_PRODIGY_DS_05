## **Finding the top features for the classification**

```python
# prompt: sklearn.featureselection chi and fscore and pscore

from sklearn.feature_selection import chi2, f_classif

# Assuming 'Accident_severity' is your target variable and you've already
# performed label encoding as shown in your previous code.

X = df.drop('Accident_severity', axis=1)  # Features
y = df['Accident_severity']  # Target variable

# Chi-squared test for non-negative features
chi2_scores, p_values = chi2(X, y)

# ANOVA F-value for numerical features
f_scores, f_p_values = f_classif(X, y)

# Create DataFrames to store the results
chi2_results = pd.DataFrame({'Feature': X.columns, 'Chi2': chi2_scores, 'P-value': p_values})
f_results = pd.DataFrame({'Feature': X.columns, 'F-score': f_scores, 'P-value': f_p_values})

print("Chi-squared test results:")
print(chi2_results)

print("\nF-value and p-value results:")
f_results
```

```python
#less important features are dropped from the df
X=X.drop(columns=['Type_of_vehicle', 'Owner_of_vehicle', 'Road_surface_conditions','Pedestrian_movement','Time', 'Day_of_week', 'Sex_of_driver','Educational_level',
                'Road_allignment', 'Sex_of_casualty', 'Casualty_severity'])
```

```python
#importing SMOTE
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
xo,yo=oversample.fit_resample(X,y)
```

## **Spliting the Dataset**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## **MOdel training and Evaluation**

```python
#KNN model using for the sample data fd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
#printing the confusion matrix

print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
